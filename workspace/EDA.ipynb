# Data Set Contentü§î
Here, I have extracted data related to 10k courses which come under the development category on Udemy's website. The 17 columns in the dataset can be used to gain insights related to:


**id**: The unique identifier assigned to each course in the dataset.

**title**: The title or name of the course as listed on Udemy's platform.

**url**: The URL of the course on Udemy's website.

**is_paid**: A boolean value indicating whether the course is paid (True) or free (False).

**num_subscribers**: The number of individuals who have subscribed to the course.

**avg_rating**: The average rating of the course based on user reviews.

**avg_rating_recent**: The recent changes in the average rating of the course.

**num_reviews**: The total number of reviews or ratings that the course has received.

**is_wishlisted**: Indicates whether the course is wishlisted by users (True) or not (False).

**num_published_lectures**: The total number of lectures available in the course.

**num_published_practice_tests**: The number of practice tests included in the course.

**created**: The timestamp indicating the creation time of the course.

**published_time**: The timestamp indicating the time when the course was published.

**discounted_price_amount**: The discounted price at which the course is being offered.

**discounted_price_currency**: The currency corresponding to the discounted price of the course.

**discounted_price_price_string**: A formatted string representing the discounted price of the course.

**price_detail_amount**: The original price of the course before any discounts.

**price_detail_currency**: The currency corresponding to the original price of the course.

**price_detail_price_string**: A formatted string representing the original price of the course.

```python
df.drop(['avg_rating', 'avg_rating_recent', 'discount_price__price_string', 'price_detail__price_string', 'discount_price__currency', 'price_detail__currency', 'url'], inplace=True, axis=1)
```

```python
df.info()
```

```python
df['created'] = pd.to_datetime(df['created'])
df['published_time'] = pd.to_datetime(df['published_time'])
```

### Changing Currency to USD

**in this step we will convert the currency from INR to USD by multiply the the actual data by 82**

```python
df.head(2)
```

```python
df['discount_price__amount'] = df['discount_price__amount'] * (1/82)
df['price_detail__amount'] = df['price_detail__amount'] * (1/82)
```

### Percent of discount
**in this step we will calculate the percentage of disconted price using the difference between the total amont and the dicounted amont column, and then we will impute the missing values in each column**

```python
df['Discount_Percentage'] = ((df['price_detail__amount'] - df['discount_price__amount']) / df['price_detail__amount']) * 100
df['Discount_Percentage'] = 100 - df['Discount_Percentage']
```

```python
imputer = SimpleImputer(strategy='median')
columns_to_impute = ['discount_price__amount', 'Discount_Percentage', 'price_detail__amount']
df[columns_to_impute] = imputer.fit_transform(df[columns_to_impute])
```

```python
df['title'] = df['title'].str.lower()
df['title'] = df['title'].str.strip().replace(r'\s+', ' ', regex=True)
df['title'] = df['title'].str.replace(r'[^\w\s]', '', regex=True)
```

### Categorizing our Courses
**In this step, we will group each course type into its specific category.**

```python
def categorize_title(title):
    title_lower = title.lower()
    if 'sql' in title_lower or 'mysql' in title_lower or 'database' in title_lower:
        return 'Database'
    elif 'tableau' in title_lower or 'power bi' in title_lower or 'data viz' in title_lower:
        return 'Data Visualization'
    elif 'excel' in title_lower or 'spreadsheet' in title_lower:
        return 'Spreadsheet'
    elif any(kw in title_lower for kw in ['agile', 'scrum', 'pmp', 'project management']):
        return 'Project Management'
    elif any(kw in title_lower for kw in ['financial', 'finance', 'accounting']):
        return 'Finance'
    elif 'mba' in title_lower or 'business' in title_lower or 'enterprise' in title_lower:
        return 'Business'
    elif any(kw in title_lower for kw in ['write', 'writing', 'editorial']):
        return 'Writing'
    elif 'sale' in title_lower or 'marketing' in title_lower:
        return 'Sales/Marketing'
    elif any(kw in title_lower for kw in ['data science', 'analytics', 'machine learning']):
        return 'Data Science'
    elif 'management' in title_lower:
        return 'Management'
    elif 'leadership' in title_lower:
        return 'Leadership'
    elif 'communication' in title_lower:
        return 'Communication'
    else:
        return 'Other'

df['category'] = df['title'].apply(categorize_title)
```

### Normal Distribution

```python

# Select numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
rows = len(numeric_columns)

# Set up subplots
fig, axes = plt.subplots(rows, 1, figsize=(10, 5 * rows), constrained_layout=True)

# Loop through numeric columns to create histograms
for i, column in enumerate(numeric_columns):
    ax = axes[i] if rows > 1 else axes  # Handle single-row case
    sns.histplot(
        df[column],
        bins=30,
        kde=True,  # Add KDE for density estimation
        stat="density",
        ax=ax,
        color="Purple"
    )
    ax.set_title(column, fontsize=14)
    ax.set_xlabel("Value", fontsize=12)
    ax.set_ylabel("Density", fontsize=12)

# Add a global title
fig.suptitle("Normal Distribution of Numeric Features", fontsize=16)
plt.show()
```

### Growth Analysis

Assuming 'published_time' is your timestamp for when the course was published, you could see how many courses are added per year.

```python
# Set the 'published_time' column as the index and resample by year
df['published_time'] = pd.to_datetime(df['published_time'])  # Ensure 'published_time' is in datetime format
courses_growth_over_time = df.set_index('published_time').resample('Y')['id'].count()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(courses_growth_over_time.index, courses_growth_over_time.values, marker='o', color='b')

# Add titles and labels
plt.title('Courses Growth Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Courses', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

### Number of Courses and Average Number of Reviews by Category
You can count the number of courses and calculate the average number of reviews per category.



```python
courses_and_reviews_by_category = df.groupby('category').agg({'title': 'count', 'num_reviews': 'mean'})
courses_and_reviews_by_category.columns = ['Courses', 'Num of Reviews']
courses_and_reviews_by_category
```

### Discount Analysis
Calculate the average discount percentage offered per category.



```python
average_discount_by_category = df.groupby('category')['Discount_Percentage'].mean()
average_discount_by_category_df = average_discount_by_category.reset_index()
average_discount_by_category_df.columns = ['Category', 'Discount Percentage']
average_discount_by_category_df
```

### Median Price of Courses by Paid/Free Status
You can find the median price of courses, separated by whether they are paid or free.

```python
paid_free_by_category = df.groupby(['category', 'is_paid']).size().unstack(fill_value=0)

# Rename the columns for better readability
paid_free_by_category.columns = ['Free', 'Paid']

# Reset the index to make 'category' a regular column
paid_free_by_category.reset_index(inplace=True)

paid_free_by_category
```

```python
# Melt the DataFrame to long format
paid_free_melted = pd.melt(paid_free_by_category, id_vars='category', value_vars=['Free', 'Paid'],
                           var_name='Type', value_name='Count')

# Set the style of the plot
sns.set_style("whitegrid")

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    x="category",
    y="Count",
    hue="Type",
    data=paid_free_melted,
    palette="BuPu"  # Color palette for the bars
)

# Add title and labels
plt.title("Counts of Paid and Free Courses by Category", fontsize=16)
plt.xlabel("Category", fontsize=14)
plt.ylabel("Count of Courses", fontsize=14)

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
